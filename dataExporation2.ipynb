```python
import numpy
import pandas as pd
import random
import json
```

```python
# Count how many total papers there are
with open("dblp.v12.json", "r") as f:
    total_lines = sum(1 for _ in f)
print(f"Total papers: {total_lines}")
```

```python
sample_size = 10000
random_indices = set(random.sample(range(total_lines), sample_size))

sampled_papers = []

with open("dblp.v12.json", "r") as f:
    for i, line in enumerate(f):
        if i in random_indices:
            if line.strip():
                clean_line = line.strip().lstrip(',')  # remove leading comma
                try:
                    sampled_papers.append(json.loads(clean_line))
                except json.JSONDecodeError:
                    print(f"Error parsing line {i}: {line[:50]}...")
                    continue

df_sample = pd.DataFrame(sampled_papers)
print(f"Successfully loaded {len(sampled_papers)} papers")
```
